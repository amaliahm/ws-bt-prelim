## JSON to Excel 

```python
from pandas import json_normalize
import pandas as pd
import json
```

```python
file_path = "/home/amina/amina/pfe/papers/output.json"
```

```python
with open(file_path, "r") as f:
    data = json.load(f) 
```

```python
data[0]["numpages"]
```

```python
df = [json_normalize(item) for item in data]
df = pd.concat(df, ignore_index=True)
```

```python
columns = {
    "id": "",
    "Title": "info.Title",
    "Author": "info.Author",
    "Co-Authors": "",
    "nombre de citation": "",
    "Publisher": "",
    "Pages": "",
    "URL": "",
    "Year": "info.CreationDate",
    "Journal": "metadata.prism:publicationname",
    "DOI": "metadata.prism:doi",
    "Keywords": "metadata.dc:subject",
    "Level de supervision": "",
    "Dataset(s) utilisé(s)": "",
    "Metriques utilisés": "",
    "Résultats clés": "",
    "Architecture ou modèle proposé": "",
    "Méthodes utilisées": "",
    "methode de generation pseudo masks": "",
    "Approche spécifique": "",
    "Limites ou remarques importantes": "",
    "Article / Conference / Review": "",
    "Dataset availibility": "",
    "Dataset URL": "",
    "Code availibility": "",
    "Code URL": ""
}

copied_columns = {
    "Title": "info.Title",
    "Author": "info.Author",
    "Co-Authors": "metadata.dc:creator",
    "Pages": "numpages",
    "Year": "metadata.prism:coverdate",
    "Journal": "metadata.prism:publicationname",
    "DOI": "metadata.prism:doi",
    "Keywords": "metadata.pdf:keywords",
    "Article / Conference / Review": "metadata.prism:aggregationtype",
    "URL": "metadata.prism:url"
}
```

```python
copied_columns.values()
```

```python
df = df[list(copied_columns.values())].copy()
df.columns = list(copied_columns.keys())
df
```

```python
columns = list(columns.keys())
```

```python
for col in columns:  
    if col not in df.columns:
        if col.lower() == "id":
            df[col] = range(1, len(df) + 1)
        else:
            df[col] = None 

df = df[columns]
df = df.astype(str)
df = df.fillna("")
df
```

```python
file_path = "/home/amina/amina/pfe/papers/result.xlsx"
df.to_excel(file_path, index=False)
```
