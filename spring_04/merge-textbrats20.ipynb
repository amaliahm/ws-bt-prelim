Libraries

```python
import shutil
import tqdm
import os
```

Merge the two datasets

```python
image_root = "/kaggle/input/datasets/awsaf49/brats20-dataset-training-validation/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData"
text_root = "/kaggle/input/datasets/aminaghandouz/textbrats20/TextBraTSData"
output_root = "/kaggle/working/BraTs20withText"
```

```python
os.makedirs(output_root, exist_ok=True)
```

```python
def find_seg_file(files):
    for f in files:
        if "seg" in f.lower():
            return f
        if "segm" in f.lower():
            return f
    return None
```

```python
for folder in tqdm.tqdm(os.listdir(image_root)):
    img_folder_path = os.path.join(image_root, folder)
    if not os.path.isdir(img_folder_path):
        continue
    files = os.listdir(img_folder_path)
    
    # find required files
    flair_file = next((f for f in files if "flair.nii" in f.lower()), None)
    seg_file = find_seg_file(files)
    
    # find text file
    text_file = os.path.join(text_root, folder, f"{folder}_flair_text.txt")
    
    # validation
    if not flair_file or not seg_file:
        print(f"skip missing image/seg {folder}")
        continue
    
    if not os.path.exists(text_file):
        print(f"skip missing text {folder}")
        continue
    
    # create output folder
    out_folder = os.path.join(output_root, folder)
    os.makedirs(out_folder, exist_ok=True)
    
    # standardized names
    new_flair = f"{folder}_flair.nii"
    new_seg = f"{folder}_seg.nii"
    new_text = f"{folder}_flair_text.txt"
    
    # copy and rename
    shutil.copy2(os.path.join(img_folder_path, flair_file), os.path.join(out_folder, new_flair))
    shutil.copy2(os.path.join(img_folder_path, seg_file), os.path.join(out_folder, new_seg))
    shutil.copy2(text_file, os.path.join(out_folder, new_text))
```
